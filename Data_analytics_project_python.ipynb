# Exploratory Data Analysis(EDA)

```python
# Gender Distribution

df["Gender"].value_counts()
```

```python
df["Gender"].value_counts().plot(kind="bar")
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("patients")
plt.show()
```

```python
df["Medical Condition"].value_counts()
```

```python
# medical Condition Distribution
df["Medical Condition"].value_counts().plot(kind="bar")
plt.title("Medical Condition")
plt.show()
```

```python
# Admission Type
df['Admission Type'].value_counts().plot(kind='pie',autopct="%1.1f%%")
plt.title("Admission type")
plt.show()
```

```python
df["Insurance Provider"].value_counts()
```

```python
# Insurance Provider
df["Insurance Provider"].value_counts().plot(kind='bar')
plt.title("Insurance Provider")
plt.show()
```

```python
df["Test Results"].value_counts()
```

```python
# Test Result
df["Test Results"].value_counts().plot(kind="bar")
plt.title("Test Results")
plt.show()
```

```python
# age distribution
plt.hist(df["Age"],bins=20)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
#Billing amount distribution
plt.hist(df["Billing Amount"],bins=20)
plt.title("Billing Amount")
plt.xlabel("Billing")
plt.ylabel("Patient")
plt.show()
```

```python
# Top 10 Hospitals
df["Hospital"].value_counts().head(10).plot(kind="bar")
plt.title("Top Hospitals")
plt.show()
```

```python
# Top 10 Doctors
df["Doctor"].value_counts().head(10).plot(kind="bar")
plt.title("Top Doctors")
plt.show()
```
